# Canva Traffic Analysis – May 2025

This analysis explores Canva's web traffic sources for May 2025, identifying opportunities to enhance user engagement and conversion rates, particularly from social media referrals.


# 2. Data Overview

```python
import pandas as pd

# Traffic source data
data = {
    'Source': ['Direct', 'Organic Search', 'Social'],
    'Percentage': [69.44, 26.9, 3.66],
    'Visits': [844.4 * 0.6944, 844.4 * 0.269, 844.4 * 0.0366]  # in millions
}

df = pd.DataFrame(data)
df['Visits'] = df['Visits'].round(2)
df
```

# 3. Visualisation


```python
import matplotlib.pyplot as plt

# Pie chart of traffic sources
plt.figure(figsize=(8, 6))
plt.pie(df['Visits'], labels=df['Source'], autopct='%1.1f%%', startangle=140)
plt.title('Canva Traffic Sources – May 2025')
plt.axis('equal')
plt.show()

```

# 4. Insights


- **Direct Traffic** dominates, accounting for nearly 70% of visits, indicating strong brand recognition.
- **Organic Search** contributes approximately 27%, reflecting effective SEO strategies.
- **Social Media** referrals are relatively low at 3.66%, suggesting potential for growth in this channel.
